```python
c = input('오빠 숫자 입력해영:')

if int(c)%2 == 0:
  print('오빵, 짝수에영')
else:
  print('오빵, 홀수에영')
```

```python
d = input('오빵 숫자 입력해영:')

if int(d) < 0:
  print('오빵, 양수에영')
elif int(d) > 0:
  print('오빵, 음수에영')
elif int(d) == 0:
  print('오빵, 0이에영')
```
