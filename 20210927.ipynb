##2. 함수를 함수의 인자로 사용


```python
def square(x):
  return x * x
```

```python
def my_func(func, i):
  return func(i)
```

```python
my_func(aquare, 3)
```

```python
def my_map(func, arg_list):
  result = []
  for i in arg_list:
    result.append(func(i))
  return result
```

```python
my_map(square, [1,2,3,4,5])
```

```python
def simple_map(arg_list):
  result = []
  for i in arg_list:
    result.append(i * i)
  return result
```

```python
simple_square = simple_map([1,2,3,4,5])
```

```python
simple_square
```

##2-1 응용

```python
def aquare(x):
  return x * x
def cube(x):
  return x * x * x
def quad(x):
  return x * x * x * x
```

```python
def my_map(func, arg_list):
  result = []
  for i in arg_list:
    result.append(func(i))
  return result
```

```python
squares = my_map(square, [1,2,3,4,5])
squares
```

```python
cubes = my_map(cube, [1,2,3,4,5])
cubes
```

```python
quads = my_map(quad, [1,2,3,4,5])
quads
```

##함수의 결과값으로 함수를 리턴


```python
def logger(msg):
  def log_message(): #1
    print('Log: ', msg)
  return log_message

# Log: Hello
```

```python
log_hello = logger("Hello")
```

```python
type(log_hello)
print(log_hello)
```

```python
log_hello()
```

```python
del logger
```

```python
log_hello()
```

```python
try:
  print(logger)
except NameError:
  print("NameError: logger는 삭제했습니다!")
finally:
  log_hello()
```

```python
log_hello()
```

```python
print(logger)
```
