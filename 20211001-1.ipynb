```python
import random
import time

mywin = 0
comwin = 0

while True:
  num = random.randrange(1,4) # 1은 가위, 2는 바위, 3은 보
  if num == 1:
    mywin = mywin + 1
  elif num == 3:
    comwin = comwin + 1
  elif num == 2:
    print('비겻당')
  print('한승준: {0}, 컴퓨터: {1}'.format(mywin, comwin))
  time.sleep(2)
```
